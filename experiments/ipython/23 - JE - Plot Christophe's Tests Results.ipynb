```python
# get test text file names
tests = []
mypath = "../Tests_fovY50degrees_aspect16-9/"
for (dirpath, dirnames, filenames) in walk(mypath):
    tests.extend(filenames)
    break
    
print tests
```

```python
data = np.genfromtxt(mypath + tests[0], delimiter='')
linear_xA = data[:, 1]

print linear_xA
linear_yA = data[:, 2]
linear_xB = data[:, 3]
linear_yB = data[:, 4]
actual_xA = data[:, 5]
actual_yA = data[:, 6]
actual_xB = data[:, 7]
actual_yB = data[:, 8]
```

```python
# graph in screen space
def drawScreenSetup():
    pylab.rcParams['figure.figsize'] = 9, 16

    plt.axes().set_aspect('equal')
    fig = plt.gcf()
    
    # Set up plot size
    plt.axes().set_xlim(-1, 1)
    plt.axes().set_ylim(-1, 1)
    
    # Draw rule of thirds
    line_l = plt.Line2D([-0.33, -0.33],[-1, 1], c="black",alpha=0.3)
    line_r = plt.Line2D([0.33, 0.33],[-1, 1], c="black",alpha=0.3)
    line_b = plt.Line2D([-1, 1],[-0.33, -0.33], c="black",alpha=0.3)
    line_t = plt.Line2D([-1, 1],[0.33, 0.33], c="black",alpha=0.3)
    fig.gca().add_artist(line_l)
    fig.gca().add_artist(line_r)
    fig.gca().add_artist(line_b)
    fig.gca().add_artist(line_t)
    
    # Draw new interpolated positions
    plt.scatter(linear_xA,linear_yA,color='red',linewidths=0)
    plt.scatter(linear_xB,linear_yB,color='red',linewidths=0)
    plt.scatter(actual_xA,actual_yA,color='blue',linewidths=0)
    plt.scatter(actual_xB,actual_yB,color='blue',linewidths=0)
```

```python
print tests[0]
drawScreenSetup()
```
