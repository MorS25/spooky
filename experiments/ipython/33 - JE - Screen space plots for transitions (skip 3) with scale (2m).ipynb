```python
%matplotlib inline
import matplotlib
import matplotlib.pylab as pylab
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.patches as mpatches
from matplotlib import gridspec

# graph in screen space
def drawScreenSetup():
    pylab.rcParams['figure.figsize'] = 9, 16

    plt.axes().set_aspect('equal')
    fig = plt.gcf() 
    
    # Set up plot size
    plt.axes().set_xlim(0, 1)
    plt.axes().set_ylim(0, 1)
    
    # Draw rule of thirds
    line_l = plt.Line2D([0.33, 0.33],[0, 1], c="black",alpha=0.3)
    line_r = plt.Line2D([0.66, 0.66],[0, 1], c="black",alpha=0.3)
    line_b = plt.Line2D([0, 1],[0.33, 0.33], c="black",alpha=0.3)
    line_t = plt.Line2D([0, 1],[0.66, 0.66], c="black",alpha=0.3)
    fig.gca().add_artist(line_l)
    fig.gca().add_artist(line_r)
    fig.gca().add_artist(line_b)
    fig.gca().add_artist(line_t)
    
    # Draw new interpolated positions
    plt.scatter(screenPos0x,screenPos0y,color='red',linewidths=0)
    plt.scatter(screenPos1x,screenPos1y,color='blue',linewidths=0)
```

```python
def getScreenData(prevShotType):
    shotType = ""
    for d in data:
        if d['camera']['isMoving']:
            if d['camera']['prevShotType'] == prevShotType:
                scale = d['camera']['scale']
                minPoint = (1.0 - scale)/2
                screenPos0x.append((d['players'][0]['screen'][0] - minPoint)/scale)
                screenPos0y.append((d['players'][0]['screen'][1] - minPoint)/scale)
                screenPos1x.append((d['players'][1]['screen'][0] - minPoint)/scale)
                screenPos1y.append((d['players'][1]['screen'][1] - minPoint)/scale)
                screenScale.append(scale)
                shotType = d['camera']['shotType']
    
    print prevShotType + " to " + shotType
    drawScreenSetup()
```
